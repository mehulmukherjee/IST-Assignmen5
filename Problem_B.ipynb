```python
df_stock = pd.read_csv('data/AAPL.csv', parse_dates=True, index_col = "Date")


# adding Date as column and index
df_stock['Date'] = df_stock.index

df = df_stock

# adding daily interval
#df.index = p.to_period('D')d.DatetimeIndex(df.index)
df.index = df.index.to_pydatetime()
df.index
```

```python
fig = go.Figure(
    data = [go.Scatter(x = df.index, y = df_stock["High"], name = "High")]
)

fig.update_layout(
    title="Stock price(High) with Rangeslider and Rangesliders",
    xaxis_title = "Index",
    yaxis_title = "High"
)

# Add rangeslider
fig.update_xaxes(        
    rangeslider_visible = True,
    range=[df.index.min(), df.index.max()],
    rangeselector_buttons = list(
        [
            dict(count = 3,  label = "Current Period", step = "month", stepmode = "backward"),
            dict(count = 6,  label = "From 2016", step = "year", stepmode = "backward"),
            dict(count = 9, label = "From 2013", step = "year", stepmode = "backward"),
            dict(count = 12, label = "Complete Period", step = "year", stepmode = "backward"),
            dict(count = 1, label = "Current Year", step = "year", stepmode = "backward"),
            dict(step="all", label = "Refresh")                                                      
        ] 
    )
)

# show figure
fig.show()
```

```python
window_size = 50

df_rolling = df['High'].rolling(window = window_size).mean()
df_rolling = df_rolling.fillna(df_rolling.mean())

fig = go.Figure()

fig.update_layout(
    title="Stock price(High) with Rangeslider and Rangesliders",
    xaxis_title = "Index",
    yaxis_title = "High"
)

fig.add_trace(go.Scatter(x = df['Date'], y = df_rolling, name="moving average"))
fig.add_trace(go.Scatter(x = df['Date'],y = df['High'], name="original"))    

fig.show()
```
